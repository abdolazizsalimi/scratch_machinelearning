```python
import numpy as np 
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import datasets 
from matplotlib import pyplot as plt 
from LinearRegression import LinearRegression
```

```python
X , y  = datasets.make_regression(n_samples= 100 , n_features= 1 , noise=20 , random_state=42)
```

```python
X
```

```python
y
```

```python
x_train , x_test , y_train , y_test = train_test_split(X,y , test_size =.3 , random_state= 12 )
```

```python
fig = plt.figure(figsize=(12,9))
plt.scatter(X[:, 0] , y , color = 'b' , marker='o'  , alpha= .8 , s=30)
plt.show()
```

```python
model = LinearRegression(learinig_rate=0.01 )
model.fit(x_train , y_train)
y_pred = model.predict(x_test)
```

```python
def mse_error(y_test , y_pred):
    mse = np.mean((y_test - y_pred)**2)

    return mse
```

```python
mse = mse_error(y_test , y_pred)
print(mse)
```

```python
predict_line = model.predict(X)
cmap = plt.get_cmap('viridis')
fig = plt.figure(figsize=(12,9))
print(x_train.shape , y_train.shape)
m1 = plt.scatter( x_train, y_train ,color=cmap(0.9) , s =10)
m2 = plt.scatter( x_test, y_test ,color = cmap(0.5) , s =10)
plt.plot(X , predict_line , color = 'black')
plt.show()
```

```python
import numpy as np 
from sklearn import datasets
import matplotlib.pyplot as plt 
from SVM  import SVM
import sys
sys.dont_write_bytecode = True
```

```python
X , y = datasets.make_blobs(n_samples=50 , n_features= 2 , centers= 2 , cluster_std= 1.025 , random_state= 42)
```

```python
X[:5]
```

```python
y[:5]
```

```python
y = np.where(y == 0 , -1 , 1)
y[:5]
```

```python
model = SVM()
```

```python
model.fit(X,y)
```

```python
def visualize_svm():
    def get_hyperplane_value(x, w, b, offset):
        return (-w[0] * x + b + offset) / w[1]

    fig = plt.figure(figsize=(12,9))
    ax = fig.add_subplot(1, 1, 1)
    plt.scatter(X[:, 0], X[:, 1], marker="o", c=y)

    x0_1 = np.amin(X[:, 0])
    x0_2 = np.amax(X[:, 0])

    x1_1 = get_hyperplane_value(x0_1, model.weight , model.bias, 0)
    x1_2 = get_hyperplane_value(x0_2,  model.weight , model.bias, 0)

    x1_1_m = get_hyperplane_value(x0_1,  model.weight , model.bias, -1)
    x1_2_m = get_hyperplane_value(x0_2,  model.weight , model.bias, -1)

    x1_1_p = get_hyperplane_value(x0_1,  model.weight , model.bias, 1)
    x1_2_p = get_hyperplane_value(x0_2,  model.weight , model.bias, 1)

    ax.plot([x0_1, x0_2], [x1_1, x1_2], "y--")
    ax.plot([x0_1, x0_2], [x1_1_m, x1_2_m], "k")
    ax.plot([x0_1, x0_2], [x1_1_p, x1_2_p], "k")

    x1_min = np.amin(X[:,1])
    x1_max = np.amax(X[:,1])
    ax.set_ylim([x1_min-3 , x1_max+3])
    plt.show()
```

```python
visualize_svm()
```

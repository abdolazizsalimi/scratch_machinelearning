```python
from PCA import PCA
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn import datasets
```

```python
load = datasets.load_iris()
data = load.data
target = load.target
```

```python
data[0:5]
```

```python
target[0:5]
```

```python
data.shape
```

```python
model = PCA(2)
model.fit(data)
X = model.transform(data)
```

```python
X.shape
```

```python
x1 = X[:,0]
x2 = X[:,1]
```

```python
plt.scatter(x1 ,x2 , c=target , edgecolors='none' , alpha=.8 , cmap= plt.cm.get_cmap('viridis' , 3 ) )
plt.colorbar()
plt.show()

print('shape of the dataset:',data.shape)
print('shape of the dataset after transform :',X.shape)
```
